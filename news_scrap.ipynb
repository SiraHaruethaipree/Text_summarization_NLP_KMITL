```python
main_web = 'https://www.investing.com/news/economy'
article = [{'article':'', 'content':'', 'link':''},
           {'article':'', 'content':'', 'link':''},
           {'article':'', 'content':'', 'link':''},
           {'article':'', 'content':'', 'link':''},
           {'article':'', 'content':'', 'link':''},
           {'article':'', 'content':'', 'link':''},
           {'article':'', 'content':'', 'link':''},
           {'article':'', 'content':'', 'link':''},
           {'article':'', 'content':'', 'link':''},
           {'article':'', 'content':'', 'link':''}]

ar = 0
for hlink in hrefs[:10]:
    article_link = main_web+hlink

    driver = webdriver.Chrome()
    driver.get(article_link)
    html_news = driver.page_source
    driver.quit()

    news_soup = BeautifulSoup(html_news, 'html.parser')
    header_name = news_soup.find('h1').text
    print(header_name)

    div = news_soup.find('div', {'class': 'WYSIWYG articlePage'})
    p_tags = div.find_all('p', recursive=False)

    # first_p = p_tags[0].text
    # print(first_p)

    other_p = [p.text for p in p_tags]
    other_p_text = ''.join(other_p[1:])
    print(other_p_text)

    article[ar]['article'] = article[ar]['article'] + header_name
    article[ar]['content'] = article[ar]['content'] + other_p_text
    article[ar]['link'] = article[ar]['link'] + article_link
    ar+=1
```

```python
tokenizer = T5Tokenizer.from_pretrained('t5-small',model_max_length=1000)

# Define the summarization methods
methods = {
 'T5-Transformer': lambda text: tokenizer.decode(T5ForConditionalGeneration.from_pretrained('t5-small').generate(tokenizer.encode("summarize: "+text, return_tensors="pt")
                      , max_length=100, num_beams=2, no_repeat_ngram_size=2, early_stopping=True)[0], skip_special_tokens=True)
    ,'Bart-large-cnn' : pipeline("summarization", model="facebook/bart-large-cnn")
}
```

```python
rouge_list_t5 = []
rouge_list_bartcnn = []
```

```python
def check_text_size(text, size = 500):
    countOfWords = len(text.split())
    if countOfWords > 500 :
        split_text = text.split()[:500]
        reduce_text = ' '.join(split_text)
        return reduce_text
    else :
        return text
```

```python
# T5
import requests
import evaluate


url = 'https://notify-api.line.me/api/notify'
token = 'fYwXov4Q8f12BiZTaU0uylYMaeEtgo6gZCzqwFLLV2Q'
headers = {
            'content-type':
            'application/x-www-form-urlencoded',
            'Authorization':'Bearer '+token
           }

for i in article:
    arti = i['article']
    content = i['content']
    link = i['link']
    content = check_text_size(content)

    summary = methods['T5-Transformer'](content)
    print(summary)

    # evalute
    predictions = [summary]
    references = [content]
    rouge = evaluate.load('rouge')
    resultst5 = rouge.compute(predictions=predictions,references=references)
    rouge_list_t5.append(resultst5)

    msg = arti+"\n\n"+summary+"\n\n"+link
    r = requests.post(url, headers=headers , data = {'message':msg})
    print(r.text)
```

```python
# BART CNN
import requests
import evaluate

url = 'https://notify-api.line.me/api/notify'
token = 'fYwXov4Q8f12BiZTaU0uylYMaeEtgo6gZCzqwFLLV2Q'
headers = {
            'content-type':
            'application/x-www-form-urlencoded',
            'Authorization':'Bearer '+token
           }

for i in article:
    arti = i['article']
    content = i['content']
    link = i['link']
    content = check_text_size(content)
    
    summary = methods['Bart-large-cnn'](content, max_length=100, min_length=30, do_sample=False)
    summary = summary[0]['summary_text']
    print(summary)

    # evalute
    predictions = [summary]
    references = [content]
    rouge = evaluate.load('rouge')
    resultsbrt = rouge.compute(predictions=predictions,references=references)
    rouge_list_bartcnn.append(resultsbrt)

    
    msg = arti+"\n\n"+summary+"\n\n"+link
    r = requests.post(url, headers=headers , data = {'message':msg})
    print(r.text)
```

```python
def calculate_mean_rouge(model_score):
    rouge1_list = []
    rouge2_list = []
    rougeL_list = []
    for i in model_score :
        rouge1_list.append(i["rouge1"])
        rouge2_list.append(i["rouge2"])
        rougeL_list.append(i["rougeL"])
    mean_rouge1 = np.mean(rouge1_list)
    mean_rouge2 = np.mean(rouge2_list)
    mean_rougeL = np.mean(rougeL_list)
    report_rouge = ["rouge1 :", mean_rouge1, "rouge2 :", mean_rouge2, "rougeL :", mean_rougeL]
    return report_rouge
```

```python
mean_t5 = calculate_mean_rouge(rouge_list_t5)
print(mean_t5)
```
